## Data Preprocessing

To merge the three datasets we would have to create a primary key, which can be created by concatinating the Year and Geography columns.

```python
Population_Housing = pd.merge(pd.merge(Households_Types, Dwelling_Households, on=['Year','Geography','Average Household Size']), Age_Gender, on =['Year','Geography'])
```
